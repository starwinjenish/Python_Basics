```python
#difference between class method & static method

class pc:
    cpu = "Intel"
    @classmethod
    def Change_cpu(cls,new_cpu):
        cls.cpu = new_cpu

    @staticmethod
    def try_change_cpu(new_cpu):
        cpu = new_cpu

pc.Change_cpu = "AMD"
print("AfterClass Method: ",pc.Change_cpu)

pc.try_change_cpu = "Amd"
print("After  Static Method :",pc.try_change_cpu)



    
```
