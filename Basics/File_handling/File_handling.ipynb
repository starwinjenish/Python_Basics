```python
#Simple write process
my_file = open("file.txt", "w")
my_file.write("Hello,This is my fisrt file handling practice Text File \n")
my_file.write("say! Allwin thayloi! then you have win 10 serrupadi from allwin\n")
my_file.close()
```

```python
#Simple append method
my_file = open("file.txt", "a")
my_file.write("Adding a new line.\n")
my_file.close()

```

```python
#with block usage - code closed automatically

with open("file.txt", "r") as my_file:
    for line in my_file:
        print(line.capitalize())
```
